```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
MESS_DATUM_formatted = pd.date_range(start = '2016-03-22 00:00:00', end = '2017-09-22 23:00:00', freq = 'H')
```

```python
data.MESS_DATUM = MESS_DATUM_formatted
```

```python
data.MESS_DATUM
```

```python
x = data.MESS_DATUM
y = data.TT_TU

# plot
plt.plot(x, y, color='red')

plt.gcf().autofmt_xdate()
plt.rcParams["figure.figsize"] = (30,10)

plt.show()
```

```python
data_august = data[(data.MESS_DATUM.dt.month == 8)]
data_august_2016 = data[(data.MESS_DATUM.dt.month == 8) & (data.MESS_DATUM.dt.year == 2016)]
data_august_2017 = data[(data.MESS_DATUM.dt.month == 8) & (data.MESS_DATUM.dt.year == 2017)]
```

```python
data_august_2016
```

```python
#x1 = data_august_2016.MESS_DATUM
#x2 = data_august_2017.MESS_DATUM
y1 = data_august_2016.TT_TU
y2 = data_august_2017.TT_TU
x = np.arange(len(y1))
#x = data_august.MESS_DATUM
#y = data_august.TT_TU

# plot
plt.plot(x, y1, color='red')
plt.plot(x, y2, color='blue')

plt.gcf().autofmt_xdate()
plt.rcParams["figure.figsize"] = (30,10)

plt.show()
```

```python
temp_cond = 25
data_juni_2017 = data[(data.MESS_DATUM.dt.month == 6) & (data.MESS_DATUM.dt.year == 2017) & (data.TT_TU > temp_cond)]
```

```python
data_juni_2017
```

```python
y = data_juni_2017.TT_TU
x = np.arange(len(y))

# plot
plt.bar(x, y, width = 0.5, color='purple')

plt.gcf().autofmt_xdate()
plt.rcParams["figure.figsize"] = (20,5)

plt.show()
```
